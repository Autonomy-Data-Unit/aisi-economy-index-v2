# fixing months 13 and 14

these months were missing full job description this was due to the retry npz from actual months did not create the job_description field.

we created full job description for all retry set sbatch z_build_df_missing_with_desc.sbatch

```python
import numpy as np
from pathlib import Path

def inspect_npz(path):

    z = np.load(path, allow_pickle=True)

    print("\n====", Path(path).name, "====")
    print("keys:", list(z.keys()))

    for k in z.keys():

        arr = z[k]

        print(f"{k:20s} shape={arr.shape} dtype={arr.dtype}")

        if arr.dtype == object or arr.dtype.kind in {"U","S"}:

            empty = (arr == "") | (arr == None)

            n_empty = empty.sum()
            share = n_empty / len(arr)

            print(f"   empty: {n_empty:,} / {len(arr):,} ({share:.2%})")

            if n_empty > 0:
                print("   example idx:", np.where(empty)[0][:5])


inspect_npz("/projects/a5u/adu_dev/aisi-economy-index/aisi_economy_index/store/AISI_demo/stage_3/prod/llm_negative_selection/bge_large/adzuna_month13.npz")

inspect_npz("/projects/a5u/adu_dev/aisi-economy-index/aisi_economy_index/store/AISI_demo/stage_3/prod/llm_negative_selection/bge_large/adzuna_month14.npz")
```

```python
import numpy as np
import pandas as pd

base = "/projects/a5u/adu_dev/aisi-economy-index/aisi_economy_index/store/AISI_demo/stage_3/prod/llm_negative_selection"

desc_path = f"{base}/df_missing_with_desc.parquet"

month_files = [
    f"{base}/bge_large/adzuna_month13.npz",
    f"{base}/bge_large/adzuna_month14.npz"
]

print("loading description table")
df = pd.read_parquet(desc_path, columns=["job_id","job_description"])

lookup = dict(zip(df.job_id.values, df.job_description.values))

for npz_path in month_files:

    print("\nprocessing:", npz_path)

    z = np.load(npz_path, allow_pickle=True)

    job_ids = z["job_ids"]

    filled = np.array([lookup.get(jid, "") for jid in job_ids], dtype=object)

    print("filled share:", (filled != "").mean())

    out_path = npz_path.replace(".npz", "_with_desc.npz")

    np.savez_compressed(
        out_path,
        job_ids=z["job_ids"],
        job_desc=z["job_desc"],
        job_tasks=z["job_tasks"],
        domain=z["domain"],
        titles=z["titles"],
        job_ad_title=z["job_ad_title"],
        job_sector_category=z["job_sector_category"],
        job_description=filled
    )

    print("saved:", out_path)
```

```python
from pathlib import Path

base = Path("/projects/a5u/adu_dev/aisi-economy-index/aisi_economy_index/store/AISI_demo/stage_3/prod/llm_negative_selection/bge_large")

pairs = [
    ("adzuna_month13.npz", "adzuna_month13_with_desc.npz"),
    ("adzuna_month14.npz", "adzuna_month14_with_desc.npz"),
]

for old_name, new_name in pairs:

    old_path = base / old_name
    new_path = base / new_name

    backup = base / old_name.replace(".npz", "_old_without_desc.npz")

    print("renaming:", old_path.name, "->", backup.name)
    old_path.rename(backup)

    print("renaming:", new_path.name, "->", old_path.name)
    new_path.rename(old_path)

print("done")
```
